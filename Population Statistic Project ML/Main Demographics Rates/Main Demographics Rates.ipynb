# Introduction and Dataset Explanation

Dataset discription related to Malaysia population: 

- Year: This column represents the year of the recorded data.

- Crude Rate of Natural Increase: It denotes the crude rate at which the population is increasing. It is typically calculated as the difference between the crude birth rate and the crude death rate.

- Crude Birth Rate: It represents the number of live births per 1,000 individuals in the population during a specific year.

- Crude Death Rate: This column indicates the number of deaths per 1,000 individuals in the population during a particular year.

- Total Fertility Rate: It denotes the average number of children that would be born to a woman over her lifetime if she were to experience the current age-specific fertility rates.

- Neonatal Mortality Rate: It represents the number of deaths of infants under 28 days of age per 1,000 live births.

- Infant Mortality Rate: This column indicates the number of deaths of infants under 1 year of age per 1,000 live births.

- Toddler Mortality Rate: It denotes the number of deaths of children between 1 and 4 years of age per 1,000 live births.

- Maternal Mortality Rate: This column represents the number of maternal deaths per 100,000 live births.

### Multivariate Regression Analysis of Maternal Mortality Rate

```python
# Define the predictor variables (X) and the response variable (y)
X = Rates[['Crude Birth Rate', 'Crude Death Rate', 'Total Fertility Rate', 'Neonatal Mortality Rate', 'Infant Mortality Rate', 'Toddler Mortality Rate']]
y = Rates['Maternal Mortality Rate']

# Add a constant term for the intercept
X = sm.add_constant(X)

# Fit the multivariate regression model
model = sm.OLS(y, X)
results = model.fit()

# Print the summary statistics
print(results.summary())
```

# Summary 

- R-squared: The R-squared value in the results is 0.780, which indicates that approximately 78% of the variability in the maternal mortality rate is explained by the independent variables included in the model. A higher R-squared generally suggests a better fit, as it implies that a larger proportion of the dependent variable's variance is accounted for by the independent variables. In this case, an R-squared of 0.780 indicates a reasonably good fit.

- Adjusted R-squared: The adjusted R-squared value is 0.766, which is slightly lower than the R-squared value. The adjusted R-squared takes into account the number of independent variables in the model, penalizing the inclusion of irrelevant variables. It is often considered a more reliable measure when comparing models with different numbers of predictors. The fact that the adjusted R-squared is slightly lower than the R-squared suggests that adding more independent variables may not substantially improve the model's predictive power.

- F-statistic: The F-statistic tests the overall significance of the regression model. In this case, the F-statistic value is 56.73, and the associated probability (Prob (F-statistic)) is very small (2.05e-29), indicating that the model as a whole is statistically significant. A significant F-statistic suggests that at least one of the independent variables is significantly related to the dependent variable.

- Coefficients: The coefficients associated with each independent variable provide insights into the direction and magnitude of their effects on the dependent variable. A statistically significant coefficient (p-value less than 0.05) suggests a significant relationship between the independent variable and the dependent variable. However, it's important to interpret the coefficients in the context of the specific variables and their units of measurement.

- While these indicators suggest that the model has a reasonably good fit and statistically significant relationships between the independent variables and the dependent variable

- The Multivariate Regression model appears to provide better results

### Cross-validation by manually splitting the data and fitting the OLS model multiple times. 

```python
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score

# Define the predictor variables (X) and the response variable (y)
X = Rates[['Crude Birth Rate', 'Crude Death Rate', 'Total Fertility Rate', 'Neonatal Mortality Rate', 'Infant Mortality Rate', 'Toddler Mortality Rate']]
y = Rates['Maternal Mortality Rate']

# Add a constant term for the intercept
X = sm.add_constant(X)

# Define the number of folds for cross-validation
n_splits = 5

# Initialize an array to store the R-squared scores
scores = []

# Perform cross-validation
kf = KFold(n_splits=n_splits, shuffle=True)
for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    # Create and fit the multivariate regression model
    model = sm.OLS(y_train, X_train)
    results = model.fit()
    
    # Make predictions on the test set
    y_pred = results.predict(X_test)
    
    # Calculate the R-squared score
    score = r2_score(y_test, y_pred)
    scores.append(score)

# Print the cross-validation scores
print("Cross-Validation R-squared scores:", scores)
print("Mean R-squared score:", np.mean(scores))
```

# Summary 

Cross-Validation:

* R-squared Scores: The cross-validation R-squared scores range from 0.607 to 0.896 across the five folds. These scores indicate the model's performance on different subsets of the data. The scores suggest that the model has relatively good predictive accuracy and generalizability.
* Mean R-squared Score: The mean R-squared score of 0.778 is calculated by averaging the individual cross-validation R-squared scores. It provides an overall estimate of the model's predictive performance. A value close to 1 suggests a good fit and indicates that the model can explain a substantial amount of the variation in the maternal mortality rate.

* Overall, the results suggest that the multivariate regression model with the selected predictor variables has a reasonable ability to explain and predict the maternal mortality rate
